```python
# 1. Get the data ready
# 2. Pick a model/estimator (to suit your problem)
# 3. Fit the model to the data and make a prediction
# 4. Evaluate the model
# 5. Improve through experimentation
# 6. Save and reload your trained model
# 7. Putting it all together (not pictured)
```

```python
# 1.get the data ready
import pandas as pd
heart_disease = pd.read_csv("C:/Users/RAJA KANNAN/Desktop/udemy machine learning/data/heart-disease.csv")
heart_disease
```

```python
#Create X (features matrix)
X = heart_disease.drop("target", axis=1)

#Create Y (Labels)
y = heart_disease['target']
```

```python
#2.Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

# We will keep the hyper parameters
model.get_params()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model.fit(X_train, y_train);
```

```python
X_test
```

```python
y_test
```

```python
# Make a prediction
y_preds = model.predict(X_test)
y_preds
```

```python
# 4.Evaluate the model on the training data and test data
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
import numpy as np
```

```python
# 5.Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators..")
    model = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy of test set: {model.score(X_test, y_test) * 100:.2f}%")
    print("")
```

```python
#6. Save model and load it
import pickle

pickle.dump(model, open("random_forest_model.pkl", "wb"))  #wb=write binary
```
